```python
import numpy as np
from PIL import Image
import os
import pandas as pd

def load_images_from_csv(csv_file, directory):
    image_data = pd.read_csv(csv_file)
    from sklearn.preprocessing import LabelEncoder

    label_encoder = LabelEncoder()

    image_data['labels'] = label_encoder.fit_transform(image_data['labels'])

    X_images = []
    y_labels = []

    valid_files = []  # Store valid image file names

    for idx, row in image_data.iterrows():
        img_name = row['image']
        img_path = os.path.join(directory, img_name)

        if os.path.exists(img_path):  # Check if the image file exists
            label = row['labels']
            image = Image.open(img_path)
            image = image.resize((224, 224))  # Resize image to a standard size
            image_array = np.array(image)  # Convert image to numpy array
            X_images.append(image_array)
            y_labels.append(label)
            valid_files.append(img_name)
        else:
            print(f"File '{img_name}' not found. Skipping...")

    # Update the dataframe to include only valid files
    valid_image_data = image_data[image_data['image'].isin(valid_files)]

    return np.array(X_images), np.array(y_labels), valid_image_data
# directory containing the images
image_directory = '/content/drive/MyDrive/images'

# path to the CSV file containing image file names and labels
csv_file = '/content/haemful_meme_for_covid19.csv'

# Load images, labels, and get valid image data
X_images, y_labels, valid_image_data = load_images_from_csv(csv_file, image_directory)

# Print the shape of loaded data
print("Shape of X_images:", X_images.shape)
print("Shape of y_labels:", y_labels.shape)
print("Valid image data:")
print(valid_image_data.head())
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models, preprocessing, optimizers
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Load the text data
text_data =df  # Assuming your text data is in a CSV file
text_data.head()# Split the data into training and testing sets
X_train_text, X_test_text, y_train, y_test = train_test_split(text_data['clean_text'], text_data['labels'], test_size=0.2, random_state=42)
X_train_images, X_test_images, y_train, y_test = train_test_split(X_images, y_labels, test_size=0.2, random_state=42)
# Convert labels to numeric format
y_train = y_train.astype(float)
y_test = y_test.astype(float)
# Tokenize text data
tokenizer = preprocessing.text.Tokenizer(num_words=10000)
tokenizer.fit_on_texts(X_train_text)
X_train_seq = tokenizer.texts_to_sequences(X_train_text)
X_test_seq = tokenizer.texts_to_sequences(X_test_text)

# Pad sequences to ensure uniform length
max_seq_length = 100
X_train_pad = preprocessing.sequence.pad_sequences(X_train_seq, maxlen=max_seq_length)
X_test_pad = preprocessing.sequence.pad_sequences(X_test_seq, maxlen=max_seq_length)
```
